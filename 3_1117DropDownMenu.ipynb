```python
import altair as alt
input_dropdown = alt.binding_select(options=['SA', 'EU', 'AF', 'NA', 'OC', 'AS'])
selection = alt.selection_single(fields=['Continent'], bind=input_dropdown, name='Country of ')
color = alt.condition(selection,
                    alt.Color('Continent:N', legend=None),
                    alt.value('lightgray'))

alt.Chart('FIFA_1112.csv').mark_bar().encode(
    alt.X("Overall Rating:Q", bin=True),
    y='count()',
    color=color,
    tooltip='Name:N'
).add_selection(
    selection
)
```
